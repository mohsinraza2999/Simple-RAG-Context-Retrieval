```python
def dataset_text_splitter(c_size,c_0_size,dataset):
  text_splitter=RecursiveCharacterTextSplitter(chunk_size=c_size,chunk_overlap=c_0_size)
  splitted_text=text_splitter.split_text(dataset)
  return splitted_text
```

```python
split_data=dataset_text_splitter(200,20,data)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
def H_F_embed_down(embed_model_name):
  h_f_embeddings=HuggingFaceEmbeddings(model_name=embed_model_name)
  return h_f_embeddings
```

```python
model_embeddings=H_F_embed_down(embed_model_name="sentence-transformers/all-MiniLM-L6-V2")
```

```python
hf_embeddings = model_embeddings.embed_documents([chunk for chunk in split_data])
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
```

```python
index = faiss.IndexFlatL2(len(model_embeddings.embed_query("hello world")))
```

```python
index
```

```python
vector_store = FAISS(
 embedding_function=model_embeddings,
 index=index,
 docstore=InMemoryDocstore(),
 index_to_docstore_id={},
 )
vector_store.add_texts(texts=split_data)
```

```python
results = vector_store.similarity_search(
    "what is faiss in ai",
    k=2
)
for res in results:
    print(f"* {res.page_content}")
```

```python
retriever = vector_store.as_retriever(search_type="mmr", search_kwargs={"k": 1})
retriever.invoke("what is faiss in AI")
```

```python
page1 = results[1].metadata.get("page", "unknown")
retrieved_context = results[0].page_content + f"\n\n(Page {page1})"
```
